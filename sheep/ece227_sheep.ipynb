```python
import networkx as nx
import numpy as np
import matplotlib
import scipy
import time
import os
import collections
import matplotlib.pyplot as plt
```

```python
G_sheep = nx.read_graphml('../Data/sheep_dominance_weighted/weighted_sheep_hass_dominance.graphml')
```

# Plot Degree Distribution


```python
degree_sequence = sorted([d for n, d in G_sheep.degree()], reverse=True)
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())

fig, ax = plt.subplots()
plt.plot(deg, cnt, color='b')

plt.title("Degree Histogram")
plt.ylabel("Count")
plt.xlabel("Degree")
```

# Diameter and Average Shortest Path Length

```python
net_diameter = nx.algorithms.distance_measures.diameter(G_sheep)
print("The network diameter of G_sheep is: ", net_diameter)
net_shortest_path = nx.algorithms.shortest_paths.generic.average_shortest_path_length(G_sheep)
print("The average shortest path of G_sheep is: ", net_shortest_path)
```

# Plot Clustering Coefficient Histogram

```python
cluster = nx.clustering(G_sheep)
count = collections.Counter(cluster.values())
count = sorted(count.items())
coeff, freq = zip(*count)
plt.semilogy(coeff, freq)
plt.title('Clustering Coefficient Distribution')
plt.xlabel('Clustering Coefficient')
plt.ylabel('Frequency');
```

# Plot Degree Centrality Histogram

```python
net_deg_cen_g1 = nx.algorithms.centrality.degree_centrality(G_sheep)

net_deg_cen_g1_list = sorted([net_deg_cen_g1[d] for d in net_deg_cen_g1]) # sorted by key, return a list of tuples
degcen_Count_g1 = collections.Counter(net_deg_cen_g1_list)
degcen_g1, cnt3_g1 = zip(*degcen_Count_g1.items())

fig, ax = plt.subplots()
plt.plot(degcen_g1, cnt3_g1, color='b')

plt.title("Degree Centrality of the nodes Histogram")
plt.ylabel("Count")
plt.xlabel("Degree Centrality")
```

# Plot the Betweenness Centrality Histogram

```python
net_bet_cen = nx.algorithms.centrality.betweenness_centrality(G_sheep)

net_bet_cen_g1_list = sorted([net_bet_cen[d] for d in net_bet_cen]) # sorted by key, return a list of tuples
betcen_Count_g1 = collections.Counter(net_bet_cen_g1_list)
betcen_g1, cnt4_g1 = zip(*betcen_Count_g1.items())

fig, ax = plt.subplots()
plt.plot(betcen_g1, cnt4_g1, color='b')

plt.title("Betweenness Centrality of the nodes Histogram")
plt.ylabel("Count")
plt.xlabel("Betweenness Centrality")
```
